```python
# Student Club Organizer

#Step 1 : Club Membership Data
coding_club = {"Alice", "Bob", "Charlie", "David"}
robotics_club = {"Charlie", "Eve", "Frank", "Bob"}
ai_club = {"George", "Alice", "Frank", "Charlie"}

#Task 1: Find student in all three clubs
all_three = coding_club & robotics_club & ai_club
print("Members in all clubs :", all_three)
```

```python
#Task 2 : Find students who are in atleast two clubs
at_least_two = (coding_club & robotics_club) | (robotics_club & ai_club) | (ai_club & coding_club)

print("Members in at least two clubs:", at_least_two)
```

```python
# Task 3 : Who is only in AI Club
only_ai = ai_club - (coding_club | robotics_club)

print("Members only in AI Club:", only_ai)
```

```python
# Step 2 : Add Scores of students (simulate event participation)
scores = {
    "Alice": 85,
    "Bob": 72,
    "Charlie": 95,
    "David": 60,
    "Eve": 78,
    "Frank": 88,
    "George": 90
}

# Task 4: Write a function to reward students scoring above 80
def reward_students(score_dict):
    """Returns a list of students who scored above 80."""
    return [name for name, score in score_dict.items() if score > 80]

print("Rewarded Students :", reward_students(scores))
```
