```python
class Student:
    def __init__(self, sno, name, course):
        self.sno=sno
        self.name=name
        self.course=course

    def details(self):
        print('Details', self.sno, self.name, self.course)

s1 = Student(12,'Marc','python')
s1.details()
s2 = Student(13, 'Alice', 'Java')
s3 = Student(14, 'Bob', 'C++')
```
