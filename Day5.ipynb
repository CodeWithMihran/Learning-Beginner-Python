```python
class calculator:
    def addition(self,x,y):
        return x+y
    def subtraction(self,x,y):
        return x-y
    def multiplication (self,x,y):
        return x*y
    def division(self,x,y):
        if (y!=0):
            return x/y
        else:
            return("cannot divide by Zero")
        
```

```python
calci=calculator()
```

```python
cResult = calci.addition(7,5)
print("the result of addition is",cResult)
```

```python
cResult = calci.subtraction(7,5)
print("the result of subtraction is",cResult)
```

```python
cResult = calci.multiplication(7,5)
print("the result of multiplication is",cResult)
```

```python
cResult = calci.division(7,5)
print("the result of division is",cResult)
```
