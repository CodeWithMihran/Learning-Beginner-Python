```python
#Common String Methods
text = "       Python Programming   "
print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("Python","Data"))
print(text.split())
```

```python
#Searching in strings
text = "   Python Programming   "
print("Py" in text)
print(text.startswith("   Py"))
print(text.endswith("ing   "))
print("o" in text)
print(text.count("o"))
print(text.find("gram"))
```

```python
'''LISTS -  Creation and Modification
# Lists are ordered collection that are mutable,
# Meaning their contents are change.
# Lists can store different data types and are used to store rows,
values, etc.'''
marks = [88, 75, 92, 67]
names = ["Alice", "Bob", "Charlie"]
mixed = [1, "hi", 3.14, True]
print(marks)
print(names)
print(mixed)
```

```python
# Accessing and modifying list elements
print(marks[0])
marks[2] = 95
print(marks)
```

```python
# Adding/Removing Elements
marks.append(81)
print(marks)
marks.insert(1, 80)
print(marks)
marks.remove(67)
print(marks)
```

```python
########  pop() function ##########
last = marks.pop()
print(marks)
# Pop the last item (default behaviour)
fruits = ["apple", "banana", "cherry", "date"]
last_fruit = fruits.pop()
print(f"Removed item: {last_fruit}")
print(f"Updated List: {fruits}")
```

```python
# Pop an item at a specific index
second_fruit = fruits.pop(1)
print(f"Removed item: {second_fruit}")
print(f"Updated list: {fruits}")
```

```python
# Sorting and Length
marks.sort()
print(marks)
marks.sort(reverse=True)
print(marks)
print(len(marks))
```

```python
# Enumerate function
# enumerate() returns both index and value while looping
names = ["Alice", "Bob", "Charlie"]
for index, name in enumerate(names):
    print(index, name)

matrix = [[1,2], [3,4], [5,6]]
print(matrix[1][0])
```

```python
# List Comprehension
# Compact way to create a new list from an existing iterable
# [expression for item in iterable if consition]

squares = [x**2 for x in range(5)]
print(squares)
```

```python
# Filtering using list comprehension
# Example : selecting even numbers
evens = [x for x in range(10) if x % 2 == 0]
print(evens)
```

```python
# List of cubes uisng list comprehension
cubes = [x**3 for x in range(5)]
print(cubes)
```

```python
# Odd number lists using filtering
odds = [x for x in range(10) if x % 2 != 0]
print(odds)
```
