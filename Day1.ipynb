```python
num1 = int(input("Enter a first number : "))
num2 = int(input("Enter a second number : "))
equal = num1 == num2
not_equal = num1 != num2
greater_than = num1<num2
greater_than_equal = num1>=num2
less_than = num1 < num2
less_than_equal = num1<=num2

# Printing the outputs
print("\nResults of Comparison Operations : ")
print(f"Is {num1} equal to {num2}? {equal}")
print(f"Is {num1} not equal to {num2}? {not_equal}")
print(f"Is {num1} grater than {num2}? {greater_than}")
print(f"Is {num1} less than {num2}? {less_than}")
print(f"Is {num1} greater than equal to {num2}? {greater_than_equal}")
print(f"Is {num1} less than equal to {num2}? {less_than_equal}")
```
