```python
"""
5.⁠ ⁠The Footprints: Sets (Finding Unique Clues)

Status: The forensic team found 10 muddy footprints leading away from the body.

Task: We don't care about all 10 prints. We want to know how many unique people were there.

Concept: A Set is a collection of Unique items. It automatically removes duplicates.
"""
```

```python
#Step 4:Analyze the foorprint(sets)
#the raw data from the floor(Size found)
raw_footprints=[10.5,9.0,10.5,10.5,9.0,12.0,10.5]
#convert to set to find unique sizes
unique_sizes=set(raw_footprints)
print(f"Raw Footprint Count:{len(raw_footprints)}")
print(f"Unique Suspects (sizes):{unique_sizes}")
#insight: we are looking for suspects with size 9.0, 10.5 or 12.0
```

```python
"""
6.⁠ ⁠The Evidence Locker: Dictionaries (Key-Value Pairs)

Status: We found items at the scene. We need to link each item to where it was found.

Task: Use a Dictionary to map items to locations.

"""
```

```python
#step 5: Bagging evidence(Dictionary)
evidence_bag={
    "weapon":"Dagger",
    "fingerprints":"Keyboard",
    "blood_stain":"carpet"
}
#what weapon was used
print(f"Weapon found:{evidence_bag["weapon"]}")
#new evidence:we found a note : add it to dictionary
evidence_bag["note"]="You will pay for this!"
print("Full Evidence Log:",evidence_bag)
```

```python
"""
7.⁠ ⁠The Interrogation: Loops & Conditionals

Status: It's time to grill the suspects. We are looking for someone with Size 10.5 shoes (from our unique_sizes analysis).

Task: Loop through the suspects. If their size matches, mark them as a Prime Suspect.

"""
```

```python
#step 6: Interrogation (Loops and logic)
#dictionary of suspects and their shoe size
suspect_profiles={
    "Saksham":11.0,
    "Chinki":8.5, 
    "Divyanshu":10.5,
    "Amit Pandey":10.5,
    "Ashish Kumar":9.0
}
print("Matching Shoe Size")
#loop --- go through name and size
for name,size in suspect_profiles.items():
    #Logic : Does the size match the crime scene evidence(10.5)?
    if size==10.5:
        print(f"Match Found:{name} is a prime suspect!")
    else:
        print(f"Cleared:{name} (Size Mismatch)")
```

```python
"""
8.⁠ ⁠The Forensic Lab: NumPy (Matrix Operations)

Status: We recovered a corrupted security camera image of the killer's face. Task: Use NumPy to enhance the image (Matrix Math).
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
#step 7: decoding the image (numpy)
#the curropted image( 5x5 matrix of pixel values)
cctv_image=np.array([
    [0,0,50,0,0],
    [0,50,255,50,0],
    [50,255,255,255,50],
    [0,50,255,50,0],
    [0,0,50,0,0]
])
print(cctv_image)
#inhance the image!
#multiply brightnesss by 2 to see clearly
enhanced_image=cctv_image*2
print("\n",enhanced_image)
#Cap value at 255(white)
enhanced_image[enhanced_image>255]=255
#Visualize
plt.imshow(enhanced_image,cmap='gray')
plt.title("Enhanced cctv evidence:")
plt.axis('off')
plt.show()
#It look like a diamond shape...
# Amit pandey is known to wear a daimond tie pin
```

```python
"""
9.⁠ ⁠The Financial Trail: Pandas (DataFrames)

Status: We have the victim's bank records. Task: Use Pandas to find suspicious transactions > $10,000,

"""
```

```python
import pandas as pd
#step 8: following the money(pandas)
data = {
    'Transaction_ID': [101, 102, 103, 104],
    'Sender': ['Parag Jain', 'Parag Jain', 'Unknown', 'Parag jain'],
    'Receiver': ['Electric Co', 'Chinki', 'Parag jain','Amit Pandey'],
    'Amount':[150, 2000, 50000, 1000000],
    'Time': ['09:00', '12:30', '18:00', '21:45']
}
bank_records=pd.DataFrame(data)
print("Victim's Financial History")
print(bank_records)
#Filter : show me transacion greater than 10000$
suspicious_tx=bank_records[bank_records['Amount']>10000]
print("\nSmoking Gun Found")
print(suspicious_tx)
```
