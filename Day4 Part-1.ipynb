```python
# Block of exception handling

# try (run this block to check exception)
# except (run this block of code when exception occurs)
# else (run this block of code when no exception occurs)
# finally (run this block of code to clean up the process)
```

```python
import math
import sys
try:
    print(math.exp(1000))
except OverflowError:
    print(sys.exc_info())
else:
    print("No Exception Occured")
```

```python
import math
import sys
try:
    print(math.exp(10))
except OverflowError:
    print(sys.exc_info())
else:
    print("No Exception Occured")
```

```python
import math
import sys
try:
    print(math.exp(1000))
except OverflowError:
    print(sys.exc_info()[1])
else:
    print("No Exception Occured")
```

```python
import math
import sys
try:
    print(math.exp(1000))
except OverflowError:
    print(sys.exc_info())
else:
    print("No Exception Occured")
finally:
    print("Program is completed. Clean the memory")
```
