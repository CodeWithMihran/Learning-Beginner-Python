```python
import matplotlib.pyplot as plt
import numpy as np
print("Matplotlib imported succesfully")
```

```python
#line plot (A line plot is used to show data trends over intervals (time, distance, etc.)
x=np.linspace(0,10,100)
y=np.sin(x)          #sin of x
plt.plot(x,y)
plt.title("Sin Wave")
plt.xlabel("X values")
plt.ylabel("Sin(x)")
plt.show()
```

```python
#Bar Chart (A bar chart is used to compare quantities across various categories)
cat=['A','B','C','D','E','F','G']
val=[4,7,1,8,5,6,7]
plt.bar(cat,val,color='pink')
plt.title("Category Values")
plt.show()
```

```python
#Random Scatter
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y,color='black')
plt.title("Random scatter plot")
plt.show()
```

```python
#Histogram (A histogram is used to show the distribution of numerical data)
data=np.random.randn(1000)
plt.hist(data,color='pink')
plt.title("Histogram")
plt.show()
```

```python
#Introduction to seaborn
import seaborn as sns
df=sns.load_dataset('tips')
df.head()
```

```python
#seaborn line plot           (A line plot is seaborn automatically handles legends,colors and styles.)
sns.lineplot(x='total_bill',y='tip',data=df)
plt.title("Line plot of total bill vs tip")
plt.show()
```

```python
#Seaborn bar plot  (A bar plot is seaborn can show averages and also confidence intervals by default)
sns.barplot(x='day',y='total_bill',hue='day',data=df,palette='magma',legend=False)
plt.title("Average Total Bill By Day")
plt.show()
```

```python
#Seaborn scatter plot          (Scatter plot is seaborn support color encoding (hue) easily)
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=df)
plt.title("Scatter plot with hue")
plt.show()
```

```python
#Seaborn distribution plot          (Distribution plot shows how a variable's values are spread out)
sns.histplot(df['total_bill'],color='green')
plt.title("Distibution of total bill")
plt.show()
```

```python
#Seaborn Box Plot (Shows Distribution and identifies outliers)
sns.boxplot(x='day',y='total_bill',data=df,color='pink')
plt.title("Box plot of total bill by day")
plt.show()
```

```python
#Seaborn pair plot (A seaborn pair plot shows pairwise relationships between numerical columns)
sns.pairplot(df)
plt.show()
```

```python
#Seaborn heatmap
#A Heatmap visualizes data in a grid where color represents values
corr=df.corr(numeric_only=True)              #corr -> co relation
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
#Create random dataset of x and y values
np.random.seed(42)
x=np.random.randint(1,50,20)
y=np.random.randint(1,50,20)
#Using matplotlib
#Line plot
plt.plot(x,y,marker='o',color='blue')
plt.title("Line plot of x vs y")
plt.xlabel("X value")
plt.ylabel("Y values")
plt.show()
#Scatter plot
plt.scatter(x,y,color='red')
plt.title("scatter plot of x vs y")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
#Histogram of x values
plt.hist(x,bins=8,color='green',alpha=0.7)
plt.title("Histogram")
plt.xlabel("X Values")
plt.show()
```

```python
# using seaborn
df=pd.DataFrame({
    'X':x,
    'Y':y,
    'Category':np.random.choice(['A','B'],size=20)
})
#Bar plot
sns.barplot(x='Category',y='Y',data=df,color='skyblue')
plt.title("Average Y by Category")
plt.show()
#Box plot
sns.boxplot(x='Category',y='Y',data=df,color='pink')
plt.title("Box plot")
plt.show()
#Heatmap
corr=df.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```
