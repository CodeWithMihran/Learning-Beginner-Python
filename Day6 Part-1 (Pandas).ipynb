```python
"""Create a dataframe with student name,subject,and score and add a column grade based on score(>90=A,>80=B,else c)fill any missing score with average group by subject and get average score sort all students by score in descending order"""
```

```python
import pandas as pd
import numpy as np
detail={
    'Name':['Amrit','Ashish','Tarun','Saksham'],
    'Subject':['Physics','Chemistry','Math','IoT'],
    'Score':[95,82,np.nan,76]
}
det=pd.DataFrame(detail)
print("Final DataFrame:")
print("\n")
print(det)

#Fill missing score with average
det['Score']=det['Score'].fillna(det['Score'].mean())

#Assign Grade
det['Grade']=det['Score'].apply(lambda x:'A' if x>90 else('B' if x>80 else 'C'))

#Group by subject
print("\n")
print("Grouped by subject:")
print("\n")
grp=det.groupby('Subject')['Score'].mean()
print(grp)

#Sort by marks
print("\n")
print("Sorted by marks:")
print("\n")
print(det.sort_values('Score',ascending=False))
```

```python
#Normal function without lambda
def assign_grd(score):
    if score>90:
        return 'A'
    elif score>80:
        return 'B'
    else:
        return 'C'
det['Grade1']=det['Score'].apply(assign_grd)
print(det)
```
